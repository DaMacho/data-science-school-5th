### 문제2: valuecount 값을 새로운 column으로 넣고 싶은데, 루프를 이용해 구현은 했으나 너무 무겁게 느껴짐(속도 저하)

```python
# 먼저 모델의 종류별 count를 파악
data['Model'].value_counts()

# 원래는 data['Frequencies'] = data['Model'].value_counts()로 시도했으나 원하는 작동이 구현 안되어 루프를 사용

# 각 모델을 일단 dictionary로 만든후 루프를 통해 나온 값 만큼 value에 +1
from collections import defaultdict
freq = defaultdict(int)
for i in data['Model']:
    freq[i] += 1
print len(freq)

# Dictionary의 Key 값과 data의 model을 매칭시켜 새로운 컬럼 Frequencies에 해당 모델이 이 dataset에서 얼마나 있는지 카운트한 숫자를 기입
data['Frequencies'] = 'none'
for i in data['Id']:
    data['Frequencies'][i] = freq[data['Model'][i]]
data.head()

# 그러나 너무 속도가 느리고 원초적인 코드라 pandas를 이용해 더욱 쉽게 할 수 있는지 궁금합니다. 
```

```python
data['Model'].value_counts()
```

```python
# group by는 나중에 심화 과정에서 다룰 것입니다.
data['Model_cnt'] = data.groupby('Model', as_index=False)['Model'].transform(lambda s: s.count())
```
