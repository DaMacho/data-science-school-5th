```python
import random

# puzzle을 아래와 같이 보기좋은 모양으로 출력함
# 1 2 3
# 4 5 6
# 7 8

def show_puzzle(puzzle):
    for i in puzzle:
        print i

# 퍼즐을 생성한다.
# 퍼즐은 동적으로 생성되며, 2차원 리스트 형태를 갖는다
# 즉 [[1, 2, 3], [4, 5, 6], [7, 8, '']]
# 빈공간은 ''로 표현한다

puzzle = [] # 함수 밖에서 puzzle 변수를 만들어 주지 않으면 안된다!!!!!
def initiate_puzzle(size):
    num_lst = range(1,(size * size))
    num_lst.append('')
    temp = []
    i = 0
    for n in range(size):
        while i < size:
            temp.append(num_lst.pop(0))
            i += 1
        puzzle.append(temp)
        temp = []
        i = 0
    return puzzle

# 퍼즐을 랜덤하게 섞음
def shuffle_puzzle(puzzle):
    dxs = [1, 0, -1,  0]
    dys = [0, 1,  0, -1]

    cnt = 0
    while cnt <= 300:
        rnd = random.randint(0, 3)
        dx = dxs[rnd]
        dy = dys[rnd]
        
        i, j = get_index(puzzle, '')
        
        ni = i + dx
        nj = j + dy

        if 0 <= ni < len(puzzle) and 0 <= nj < len(puzzle[0]):
            puzzle[ni][nj], puzzle[i][j] = puzzle[i][j], puzzle[ni][nj]
        cnt += 1

# 퍼즐이 종료되었는지, 즉 차례대로 정렬되어 완료가 되었는지 검사
def is_done(puzzle, complete):
    if puzzle == complete:
        return True

# 퍼즐 보드에서 숫자 n의 인덱스 검색
# 2차원이기 때문에 i, j형태의 인덱스 값을 반환(j가 x, i가 y)
def get_index(puzzle, n):
    idx = []
    for y in puzzle:
        for x in y:
            if x == n:
                idx.append(y.index(x))
                idx.append(puzzle.index(y))
    return idx
    #if idx:
        #return idx
    #else:
        #print "You've selected Incorrect number"
                
                

# 퍼즐에서 숫자 n을 이동,
# 이동할 수 없는 경우에는 이동할 수 없다고 표시
# 이동이 가능한 경우는 양옆위아래에 ''가 위치해 있을 경우이다
def move_by_number(puzzle, n):
    x, y = get_index(puzzle, n)
    move_by_index(puzzle, x, y)
 
    

# 숫자를 이동시키기 위해서는 결국 인덱스를 알아야 함
# 즉 move_by_number 내부에서 호출되는 함순
def move_by_index(puzzle, x, y):
    # movable condition
    # 좌우위아래 한방향중 하나가 '' 값이라면 이동 가능
    for mx, my in [[-1, 0], [1, 0], [0, -1], [0, 1]]:
        if 0 <= x + mx < len(puzzle) and 0 <= y + my < len(puzzle):
            if puzzle[y+my][x+mx] == '':
                puzzle[y+my][x+mx], puzzle[y][x] = puzzle[y][x], puzzle[y+my][x+mx]
                break
    else:
        print "We could not move the number"
  


# 퍼즐 생성
size = int(raw_input(' -> please insert puzzle size : '))
puzzle = initiate_puzzle(size)

# 연습문제) -> 아래와 같이 복사하는 이유는?
complete = [row[:] for row in puzzle]

# 퍼즐 섞기
shuffle_puzzle(puzzle)

# 섞인 퍼즐 보기
show_puzzle(puzzle)

# 퍼즐 풀기
while not is_done(puzzle, complete):
    num = raw_input(' -> select a number to move : ')
    # 움직일 수 선택하기
    if num == 'q':
        print "OK, Try next time"
        break
    else:
        move_by_number(puzzle, int(num))

    # 화면 clear
    import os
    cls = lambda: os.system('clear') # windows cls
    cls()

    # 움직인 이후 퍼즐 상태 보기
    show_puzzle(puzzle)

print '\nYou solved the puzzle!'
```
