# 재귀함수 크롤링

```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
# depth 3으로 했을때 에러 발생. depth 2로 한 결과로 대체
# 하단에 depth 3에서의 에러 포함 

def re_url(url, depth):
    if depth >= 2:
        return url

    req = requests.get(url)
    soup = BeautifulSoup(req.text, 'html5lib')

    url_in = soup.select('a[href]')   

    for link in url_in:
        if re.search(r'https?://\w+', link['href']):
            if link['href'] not in url_list:                
                url_list.append(link['href'])
                print re_url(link['href'], depth + 1)
            
                        
depth = 0  

url_list= []

urls = 'http://v.media.daum.net/v/20170626133016145'

re_url(urls, depth)
```

```python
# depth 3에서는 오류 발생. depth 2로는 오류 없이 동작함

def re_url(url, depth):
    try:
        if depth >= 3:
            return url
       
        req = requests.get(url)
        soup = BeautifulSoup(req.text, 'html5lib')
    
        url_in = soup.select('a[href]')   
    
        for link in url_in:
            if re.search(r'https?://\w+', link['href']):
                if link['href'] not in url_list:                
                    url_list.append(link['href'])
                    print re_url(link['href'], depth + 1)
    except UnicodeEncodeError as e:
        print e           
                
                        
depth = 0  

url_list= []

urls = 'http://v.media.daum.net/v/20170626133016145'

re_url(urls, depth)
```
